The present notebook graphs the ReLU, sigmoid, tanh and Softplus activation functions used in Machine Learning. Each function is graphed separately

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-10, 10, 100)
```

```python
functions = pd.DataFrame()
functions
```

```python
functions["ReLU"] = [x if x >=0 else 0 for x in x]
functions["sigmoid"] = [1/(1+math.exp(-x)) for x in x]
functions["tanh"] = [math.tanh(x) for x in x]
functions["softplus"] = [math.log(1 + math.exp(x)) for x in x]
```

```python
functions.index = x
```

```python
plt.plot(functions["ReLU"])
plt.grid()
plt.title("ReLU Activation Function")
```

```python
plt.plot(functions["sigmoid"])
plt.grid()
plt.title("Sigmoid Activation Function")
```

```python
plt.plot(functions["tanh"])
plt.grid()
plt.title("tanh Activation Function")
```

```python
plt.plot(functions["softplus"])
plt.grid()
plt.title("Softplus Activation Function")
```
